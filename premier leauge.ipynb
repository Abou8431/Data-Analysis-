```python
# 1. Group by team to get total wins per team
most_wins = df.groupby('team')[['wins']].sum()

# 2. Get the names of the top 5 teams with most wins
top_5_teams = most_wins.sort_values('wins', ascending=False).head(5).index

# 3. Filter original df to include only rows for top 5 teams
df_top5 = df[df['team'].isin(top_5_teams)]

# 4. Define goal-related stats
goal_related_stats = [
    'att_hd_goal', 'att_pen_goal', 'att_freekick_goal',
    'att_ibox_goal', 'att_obox_goal', 'goal_fastbreak',
    'own_goals', 'pen_goals_conceded'
]

# 5. Group by team and sum the goal-related stats
team_goals_summary = df_top5.groupby('team')[goal_related_stats].sum()

# Area plot (transposed so that stats are on the x-axis)
team_goals_summary_T = team_goals_summary.T

team_goals_summary_T.plot(kind='area', figsize=(12, 6), alpha=0.6)

plt.title('Goal-Related Stats (Area Plot) for Top 5 Teams by Wins')
plt.xlabel('Stat')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.legend(title='Team')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
 #Get top 5 and bottom 5 teams by wins
top_5 = df.sort_values(by='wins', ascending=False).head(5)
bottom_5 = df.sort_values(by='wins', ascending=True).head(5)

# Defensive features
defensive_features = ['goals_conceded', 'clean_sheet', 'interception', 'clearance_off_line', 'outfielder_block', 'own_goals', 'head_clearance','dispossessed','total_offside','total_tackle']

# Compute averages
top5_def = top_5[defensive_features].sum()
bottom5_def = bottom_5[defensive_features].sum()

# Combine for comparison
comparison_df = pd.DataFrame({'Top 5 Teams': top5_def, 'Bottom 5 Teams': bottom5_def})

# Plot
comparison_df.plot(kind='bar', figsize=(8,6), title="Defensive Stats: Top vs Bottom 5 Teams", rot=20)
plt.ylabel("Total Value")
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Create a binary target column for high-win vs low-win teams
median_wins = df['wins'].median()
df.loc[:, 'win_label'] = (df['wins'] >= median_wins).astype(int)
```

```python
# Step 2: Prepare feature and target sets
X = df.drop(columns=['team', 'season', 'wins', 'win_label'])  # Drop non-numeric or non-useful
y = df['win_label']

# Step 3: Fill missing values with column means
X = X.fillna(X.mean())

# Step 4: Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 5: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 6: Train Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=0)
rf.fit(X_train, y_train)

# Step 7: Predict and evaluate
y_pred = rf.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Step 8: Feature Importance
feature_importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
top_features = feature_importances.head(10)

# Step 9: Plot Feature Importance
plt.figure(figsize=(10,6))
sns.barplot(x=top_features.values, y=top_features.index, palette='viridis')
plt.title("Top Features Influencing Wins")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Train Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Predict and Evaluate
y_pred_log = log_reg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
```
